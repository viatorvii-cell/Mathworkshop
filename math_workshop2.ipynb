Question no 1

```python
def unique_list(lst):
    new_list = []
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    return new_list

sample = [1, 2, 3, 3, 3, 3, 4, 5]
print("Unique List:", unique_list(sample))
```

Question no 2

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print("Prime number")
else:
    print("Not a prime number")
```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

num = 7
while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number:", num)
        break
    num += 1
```

Question no 6

```python
marks = [20, 30, 35, 55, 70, 80, 90]

total = 0
for m in marks:
    total += m
mean = total / len(marks)

n = len(marks)
if n % 2 == 1:
    median = marks[n // 2]
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2

print("Mean:", mean)
print("Median:", median)
```

Question no 7

```python
def decimal_to_binary(n):
    binary = ""
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary

num = float(input("Enter decimal number: "))
print("Binary:", decimal_to_binary(num))
```

Question no 8

```python
def prodDigits(num):
    product = 1
    while num > 0:
        digit = num % 10
        product *= digit
        num = num // 10
    return product

n = int(input("Enter a number: "))
print("Product of digits:", prodDigits(n))
```
