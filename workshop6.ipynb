Qno 8

```python
def mat_vec_mult(A, x):
    result = []
    for i in range(len(A)):
        s = 0
        for j in range(len(x)):
            s += A[i][j] * x[j]
        result.append(s)
    return result

A = [[1,2,3],
     [4,5,6],
     [7,8,9]]

x = [1,1,1]

print("Manual:", mat_vec_mult(A,x))

A_np = np.array(A)
x_np = np.array(x)

print("NumPy:", A_np.dot(x_np))
```

Qno 9

```python
def mat_mat_mult(A, B):
    result = [[0]*len(B[0]) for _ in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

    return result

A = [[1,2,3],
     [4,5,6]]

B = [[1,2],
     [3,4],
     [5,6]]

print("Manual:", mat_mat_mult(A,B))

A_np = np.array(A)
B_np = np.array(B)

print("NumPy:\n", A_np.dot(B_np))
```

Qno 10

```python
def is_symmetric(A):
    A = np.array(A)
    return np.array_equal(A, A.T)

A = [[1,2,3],
     [2,5,6],
     [3,6,9]]

print(is_symmetric(A))
```
